```python
class ATM:
    def __init__(self, account_balance, pin):
        self.account_balance = account_balance
        self.pin = pin

    def check_balance(self):
        return self.account_balance

    def deposit(self, amount):
        if amount > 0:
            self.account_balance += amount
            return f"Deposited RS{amount}. New balance: RS{self.account_balance}"
        else:
            return "Invalid deposit amount."

    def withdraw(self, amount):
        if amount > 0 and amount <= self.account_balance:
            self.account_balance -= amount
            return f"Withdrew RS{amount}. New balance: RS{self.account_balance}"
        elif amount > self.account_balance:
            return "Insufficient funds."
        else:
            return "Invalid withdrawal amount."

def main():
    account_balance = 2000  
    pin = "2028"  
    atm = ATM(account_balance, pin)

    entered_pin = input("Enter your PIN: ")
    if entered_pin == atm.pin:
        while True:
            print("\nOptions:")
            print("1. Check Balance")
            print("2. Deposit")
            print("3. Withdraw")
            print("4. Exit")
            choice = input("Select an option (1/2/3/4): ")

            if choice == "1":
                print(f"Balance: RS{atm.check_balance()}")
            elif choice == "2":
                deposit_amount = float(input("Enter the deposit amount: RS"))
                print(atm.deposit(deposit_amount))
            elif choice == "3":
                withdraw_amount = float(input("Enter the withdrawal amount: RS"))
                print(atm.withdraw(withdraw_amount))
            elif choice == "4":
                print("Thank you for using this ATM. Goodbye!")
                break
            else:
                print("Invalid choice. Please select a valid option.")

    else:
        print("Invalid PIN. Access denied.")

if __name__ == "__main__":
    main()
```
